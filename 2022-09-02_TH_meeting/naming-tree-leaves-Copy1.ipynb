```python
file = 'hel1.fasta'
```

```python
d = {}

with open(file) as f:
    for line in f:
        if len(line) > 1:
            if '>' in line:
                key = line.strip()
                d[key] = ""
            else:
                d[key] += line.strip() + "+"

d = {key: d[key][:-1] for key in d}
```

```python
new = pd.DataFrame.from_dict(d, orient ='index')
```
