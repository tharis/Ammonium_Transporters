```python
your_tree = Tree("transporter_tree_renamed.new", quoted_node_names=True)
```

```python
R = your_tree.get_midpoint_outgroup()
your_tree.set_outgroup(R)
```

```python
your_tree_style = ete3.TreeStyle()
your_tree_style.mode = "c"

your_tree_style.show_branch_support = True
your_tree_style.show_leaf_name = True

your_tree.render("SCOT_0002.1.pdf", tree_style=your_tree_style) 
```

```python
def conditional_function(node):
    for node in your_tree:
        if 'Archaeoglobus' in node.name:
            nst1 = NodeStyle()
            nst1['bgcolor'] = 'blue'
            node.set_style(nst1)
        elif node.name.split(' ')[2] in Bst:
            nst3 = NodeStyle()
            nst3['bgcolor'] = 'red'
            node.set_style(nst3)
        elif node.name.split(' ')[2] in Est:
            nst4 = NodeStyle()
            nst4['bgcolor'] = 'cyan'
            node.set_style(nst4)
        elif 'metagenome' in node.name:
            nst5 = NodeStyle()
            nst5['bgcolor'] = 'green'
            node.set_style(nst5)
        else:
            nst2 = NodeStyle()
            nst2['bgcolor'] = 'yellow'
            node.set_style(nst2)
```

```python
ts = ete3.TreeStyle()
ts.layout_fn = conditional_function
ts.mode = "c"

ts.show_branch_support = True
ts.show_leaf_name = True

your_tree.render("SCOT_0002.6.pdf", tree_style=ts)
```
